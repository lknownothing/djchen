### 批量下载
加上两层循环就可以批量下载了

```python
import requests
from  bs4 import BeautifulSoup
import os
import sys
print(os.name)
```

```python
if(os.name=="nt"):
    print("You ars use Win platform\n")
else:
    print("You are use Linux platform\n")
header={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.130 Safari/537.36"}

all_url="https://www.mzitu.com"
start_html=requests.get(all_url,headers=header)
path="D:/爬虫/mzitu/"

# 寻找最大页数
soup=BeautifulSoup(start_html.text,"html.parser")
page=soup.find_all("a",class_="page-numbers")
max_page=page[-2].text
print(max_page)

same_url="https://www.mzitu.com/page"
for n in range(2,int(max_page)+1): # 对于index所有的页
    url=same_url+'/'+str(n)
    start_html=requests.get(url,headers=header)
    soup=BeautifulSoup(start_html.text,"html.parser")
    all_a=soup.find("div",class_="postlist").find_all("a",target="_blank")
    for a in all_a:   # 对应每一页中列取出的类别
        title=a.get_text() # 提取文本
        if(title !=""):
            print("准备爬取："+title)
            if(os.path.exists(path+title.strip().replace("?",""))):
                flag=1  # 目录已经存在
            else:
                os.makedirs(path+title.strip().replace("?",""))
                flag=0 # 
                
            os.chdir(path+title.strip().replace("?","")) # 更换cwd为path中的目录
            href=a["href"]
            html=requests.get(href,headers=header)
            mess=BeautifulSoup(html.text,"html.parser")
            pic_max=mess.find_all("span")
            pic_max=pic_max[9].text # 最大页数
            
            if(flag==1 and len(os.listdir(path+title.strip().replace("?",""))))>=int(pic_max):
                print("已经保存完毕，跳过")
                continue
            
            for num in range(1,int(pic_max)+1): # 每一类进去有多少页图片
                pic=href+"/"+str(num)

                html=requests.get(pic,headers=header)  # 添加header十分的关键
                
                mess=BeautifulSoup(html.text,"html.parser")
                
                pic_url=mess.find("img",alt=title)  # 图片所在的具体网址
                                     
                html=requests.get(pic_url["src"],headers=header)
                file_name=pic_url["src"].split(r'/')[-1]
                f=open(file_name,"wb")  # 因为更改过当前工作目录，所以会在更改过的dir目录下直接写入
                f.write(html.content)
                f.close()
            print("完成")
    print("第n页完成")
            
            
            
        
    
    
```

## 总结
爬取多层结构式，在外层加入一层嵌套
+ chdir(path) 修改当前的工作目录
+ 先find()再find_all()
