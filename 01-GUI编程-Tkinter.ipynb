### Tkinter编程
tkinter是Python的标准的GUI库，Python使用tkinter可以快速构建GUI应用程序。
### 创建一个GUI程序
1. 导入Tkinter
2. 创建控件
3. 指定这个控件的master,即这个控件时属于哪一个
4. 告诉GM(geometry manager)有一个控件产生了


```python
import tkinter
from tkinter import *
top=tkinter.Tk()  # 创建了一个窗口
# 进入消息循环
top.title("This is a test for top")

# 在窗口中船舰标签
label_test=Label(top,text="Hello label_test")
label_test.pack()
top.mainloop()
```

```python
from tkinter import *
root=Tk()  # 创键窗口对象的背景色
li=["C","Python","html"]
movie=["黑鹰坠落","世界末日"]

# 创键两个list组件
list1=Listbox(root)
list2=Listbox(root)

# 给两个小部件插入数据
for item in li:
    list1.insert(0,item)

for item in movie:
    list2.insert(0,item)

list1.pack()  # 将小部件放入到主窗口中
list2.pack()
root.mainloop()
```

### Tkinter组件
+ Button 按钮控件
+ Canvas 画布控件 显示图形元素，如线条或者文本
+ Checkbutton 多选框控件
+ Entry 输入控件
+ Frame 矿建控件 屏幕上显示一个矩形区域，多用来作为容器
+ Label 标签控件 可以显示文本或者位图
+ Listbox 列表框控件 在listbox窗口小部件是用来显示一个字符串列表给用户
+ Menubutton 菜单按钮控件 用于显示菜单项
+ Menu 菜单控件 显示菜单或者弹出菜单
+ Message 消息控件 用来显示多行文本，与消息类似
+ Radiobutton 单选按钮控件，显示一个单选的按钮状态
+ Scale 范围控件
+ Scrollbar 滚动控件
+ Text 文本控件
+ Toplevel 容器控件
+ Spinbox 输入控件
+ PaneWindow 窗口布局管理的插件，可以包含一个或者多个子控件
+ LabelFrame 是一个简单的容器控件
+ tkMessageBox 

### 标准属性
+ Dimension 控件大小
+ Color 控件颜色
+ Font 控件字体
+ Anchor 锚点
+ Relief 控件样式
+ Bitmap 位图
+ Cursor 光标

### 选择按钮
选择按钮时，用一个标签显示：你已经选择了按钮。
我们需要顶一个var用来将radiobutton的值和label值联系在一起。


```python
win2=tk.Tk()
var2=StringVar()
win2.title("Radionbutton Test")
win2.geometry("400x400")
l2=tk.Label(win2,bg="yellow")  # label时使用textvariable
l2.pack()

def print_selection():
    l2.config(text="You have choose "+var2.get()) # 通过config设置var的变量
r1=tk.Radiobutton(win2,text="Option A",variable=var2,value="A",
                 command=print_selection)  # value=A 鼠标选中时候，把value的值放到变量var中

r1.pack()
r2=tk.Radiobutton(win2,text="Option A",variable=var2,value="B",
                 command=print_selection)
r2.pack()
r3=tk.Radiobutton(win2,text="Option A",variable=var2,value="C",
                 command=print_selection)
r3.pack()
win2.mainloop()
    
```

### Canvas


```python
win5=tk.Tk()
win5.title("Canvas Test")
win5.geometry("400x400")
canvas=tk.Canvas(win5,bg="blue",width=200)
canvas.pack()
image_file=tk.PhotoImage(file="./ins.gif")  # 创造一个变量来存放ins.gif这个图片
image=canvas.create_image(10,10,anchor="nw",image=image_file) # 10，10是图片放入画布的坐标
# anchor="nw"，就是讲图片放入左上方，image=image_file  就是讲刚存入的图片变量给image
x0, y0, x1, y1= 50, 50, 80, 80
line = canvas.create_line(x0, y0, x1, y1)
oval = canvas.create_oval(x0, y0, x1, y1, fill='red')  # 通过四个点画一个圆
arc=canvas.create_arc(x0+30, y0+30, x1+30, y1+30, start=0, extent=180) # 创键一个扇形
rect = canvas.create_rectangle(100, 30, 100+20, 30+20)   #创建一个矩形
canvas.pack()

def moveit():
    canvas.move(rect,0,2)

b5=tk.Button(win5,text="move",command=moveit).pack()
win5.mainloop()
```

### Menubar菜单


```python

win6=tk.Tk()
win6.title("My Mene")
win6.geometry("400x400")

l6=tk.Label(win6,bg="red",text="hello Menubar")
l6.pack()

count=0
def do_job():
    global count
    l.config(text="do"+str(count))
    count+=1
    

# 创键一菜单栏，理解成一个容器
menubar=tk.Menu(win6)

# 定义一个空的菜单单元
filemenu=tk.Menu(menubar,tearoff=0)

# 将上面定义的空菜单命名为”file",放在菜单栏中，就是装入那个容器中
menubar.add_cascade(label="文件",menu=filemenu)

filemenu.add_command(label="新建",command=do_job)
filemenu.add_command(label="打开",command=do_job)
filemenu.add_command(label="保存",command=do_job)

filemenu.add_separator() # 这里就是一条分割线

filemenu.add_command(label="退出",command=window.quit)

# 同样我我们在顶一个一个和文件一样的menu
editmenu=tk.Menu(menubar,tearoff=0)
menubar.add_cascade(label="编辑",menu=editmenu)

editmenu.add_command(label="剪切",command=do_job)
editmenu.add_command(label="复制",command=do_job)

submenu=tk.Menu(filemenu) # 创键一子菜单
filemenu.add_cascade(label="Import",menu=submenu) # 装容器中
submenu.add_command(label="导出到文件",command=do_job)

win6.config(menu=menubar)

win6.mainloop()
```

### Frame框架
Frame实在Windows上分离小区的部件，它能讲windows分成不同的区，然后存放同的其它部件，同时一个fram也能分成两个frame,frame可以认为是一种仪器。


```python
win7=tk.Tk()
win7.title("Frame 部件")
win7.geometry("400x500")

fr=tk.Frame(win7)
fr.pack()

fr_l=tk.Frame(fr)
fr_r=tk.Frame(fr)
fr_l.pack(side="left")
fr_r.pack(side="right")

tk.Label(fr_l,text="这是偏左的frame").pack()
tk.Label(fr_r,text="这是一个偏右的Frame").pack()
win7.mainloop()
```
