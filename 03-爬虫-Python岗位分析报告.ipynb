什么式Ajax?
Ajax等于异步JavaScript和XML。是一种用于创建快速动态网页的技术。通过在后台与服务器进行少量的数据交换，ajax可以使得网页实现异步更新，这意味着可以不重新加载整个网页的情况下，对网页进行部分更新。传统网页（不使用ajax）如果需要更新内容，需要加载整个网页页面。
有很多使用ajax的应用程序案例：新浪微博，Google地图，开心网等。
Ajax工作原理：
+ 浏览器：发生某个事件，创建XMLHttpRequest对象，发送HttpRequest.
+ 服务器：处理HttpRequest 创建响应返回浏览器。responseText获得来自服务器的字符串的响应数据，responseXML 获得xml形式的响应数据，
    + document.getElementById("myid").innerHTML.xml.responseText
    + 如果来自服务器的响应是xml,而且需要作为xml对象进行解析。
    xmlDoc=xmlhttp.responseXML;
    txt="";
    x=getElementsByTagName("ARTIST")
    
+ 浏览器：使用JS处理被返回的数据，更新页面内容
Ajax基于现有的Internet标准：
+ XMLHttpRequest对象（异步的与服务器交换数据）
+ JavaScript/DOM (信息显示/交互）
+ CSS 给数据定义样式
+ XML 作为转换数据的格式

### Ajax onreadystatechange事件
当请求被发送到服务器时，我们需要执行一些基于响应的任务。
+ onreadystatechange 存储函数名，每当readState属性改变时，就会调用该函数。注意：上面的onreadystatechange事件被触发4次。
+ readystate每当readystate改变时，就会触发onreadystatechange事件。readystate属性存有XMLHttpRequest的状态信息。0: 请求未初始化；1:服务器链接已经建立；2：请求已经接受，3：请求处理中；4:请求已经完成，并响应已经就绪。
+ status, 200:"ok”，404:"未找到页面"
当readstate==4 且xmlhttp.status==200时，表示响应已经就绪。

### 获取所有数据
了解如何解析数据，剩下的就是连续请求所有页面了，我们构造一个函数来请求所有30页的数。



```python
import random
import time
import requests
from openpyxl import Workbook
import pymysql.cursors

# def get_conn():
#     # 建立数据库连接
#     conn=pymysql.connect(host="localhost",
#                         user="root",
#                         db="python",
#                         charset="utf8mb4",
#                         cursorclass=mysql.cursor.DictCursor)
#     return conn

# def insert(conn,info):
#     # 数据写入数据库
#     with conn.cursor() as cursor:
#         sql="INSERT INTO 'python'("shortname","fullname","industryfield","companySize","salary","city","education") VALUES(%s %s %s %s %s %s %s)"
#         cursor.execute(sql,info)


# 输入网址，页码，搜索名称 
# 目的是得到岗位的info_list
def get_json(url,page,lang_name):
    headers={
        "Host":"www.lagou.com",
         "Origin":"https://www.lagou.com", # 可行
        "Connection":"keep-alive", # 可行
#          "Content-Length":"25",
        
        "X-Anit-Forge-Code":"0",
        "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.130 Safari/537.36",
        "Content-Type":"application/x-www-form-urlencoded; charset=UTF-8",
        "Accept":"application/json, text/javascript, */*; q=0.01",
        "X-Requested-With":"XMLHttprRequest",
        "X-Anit-Forge-Token":"None",
        "X-Anit-Forge-Code":"0",
        "Referer":"https://www.lagou.com/jobs/list_python/p-city_2?px=default",
        "Accept-Encoding":"gzip,deflate, br",
        "Accept-Language":"zh-CN,zh;q=0.9"
                
    }
    data={"first":"false","pn":page,"kd":lang_name}
    # 配置完毕
    
    #　得到请求的 解析数据
    s = requests.Session()
    urls = 'https://www.lagou.com/jobs/list_Python?labelWords=&fromSearch=true&suginput='
    # 获取搜索页的cookies
    s.get(urls, headers=headers, timeout=5)
    # 为此次获取的cookies
    cookie = s.cookies
#     print(cookie)
    # 获取此次文本
    json = s.post(url, data=data, headers=headers, cookies=cookie).json()
#     json=requests.post(url,data,headers=headers).json()
#     print(type(json))
#     print(json.get("content"))  # 页面没有加载
    list_con=json['content']["positionResult"]["result"]  # result中就是一个list
    info_list=[]
    for i in list_con:
       
        info=[]
        info.append(i.get("companyShortName",'无'))
        info.append(i.get("companyFullName","无"))
        info.append(i.get("companySize","无"))
        info.append(i.get("industryField","无"))
        info.append(i.get("salary","无"))
        info.append(i.get("city","无"))
        info.append(i.get("education","无"))
        info_list.append(info) # 把info中的信息都加到info_list中
#         print(info_list)
    return info_list   
        
def main():
    lang_name="python"
    wb=Workbook()
#     conn=get_conn()  # 建立数据库，不存数据库 注释此行
    for i in ["北京","上海","广州","深圳","杭州"]:
#     for i in ["北京","上海"]:
        page=1
        ws1=wb.active  # 第一个sheet
        ws1.title=lang_name # sheet的名字
        url="https://www.lagou.com/jobs/positionAjax.json?city={}&needAddtionalResult=false".format(i)
        while page <18:
            info=get_json(url,page,lang_name)
            page+=1
            time.sleep(random.randint(10,20))
            for row in info:
                ws1.append(row)
    wb.save("{}职位信息.xlsx".format(lang_name))

if __name__=="__main__":
    main()
                
            
        
```
