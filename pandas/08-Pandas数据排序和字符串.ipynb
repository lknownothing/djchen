### Series的排序
+ Series.sort_values(ascending=True,inplace=False)
+ DataFrame.sort_values(by,ascengding=True,inplace=False)
    + by: 字符串或者list,单列或者多列排序

```python
import pandas as pd
```

```python
fpath = "./datas/beijing_tianqi/beijing_tianqi_2018.csv"
df = pd.read_csv(fpath)

# 替换掉温度的后缀℃
df.loc[:, "bWendu"] = df["bWendu"].str.replace("℃", "").astype('int32')
df.loc[:, "yWendu"] = df["yWendu"].str.replace("℃", "").astype('int32')
```

```python
df.head()
```

```python
df["bWendu"].sort_values(ascending=False,inplace=False) # 是一个view,不能在原位置排序
```

```python
df.sort_values(by=["fengxiang","aqi"])  
```

```python
df.sort_values(by=["aqi","fengxiang"])
```

### 转换为中文日期


```python
def func(x):
    y,m,d=x["ymd"].split("-")  # 返回有三个元素的,已经是str的就不用str属性了
    return f"{y}年{m}月{d}日"
df["中文日期显示"]=df.apply(func,axis=1)
df
```

### series默认就开启了正则表达式
