### 实现excel的vlookup并在指定列后边输出
+ vlookup: vlookup(查找值，查找范围，查找列数，false) 返回的是查找列数的值 

案列要求：  
新增的列要放在excel指定位置的后边

```python
import pandas as pd
# 学生成绩表
df_grade = pd.read_excel("./course_datas/c23_excel_vlookup/学生成绩表.xlsx") 
df_grade.head()
```

```python
df_grade["班级"].value_counts()  # 只有一个班的数据
```

```python
df_sinfo = pd.read_excel("./course_datas/c23_excel_vlookup/学生信息表.xlsx") 
df_sinfo.head()
```

```python
df_merge=pd.merge(df_grade,df_sinfo,left_on="学号",right_on="学号",how="outer")
df_merge
```

```python
# 如何将姓名，性别放在学号后边
new_col=df_merge.columns.to_list()
print(new_col)
print(new_col.index("学号"))
```

```python
for name in ["性别","姓名"]:
    new_col.remove(name)
    new_col.insert(new_col.index("学号")+1,name)
print(new_col)
```

```python
df_merge=df_merge.reindex(columns=new_col)
df_merge.head()
```

```python
import os
root_path="./datas/after_deal"
if not os.path.exists(root_path):
    os.mkdir(root_path)
df_merge.to_excel(f"{root_path}/vookup.xlsx",index=False)
```
