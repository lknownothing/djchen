```python
import pandas as pd
fpath = "./datas/beijing_tianqi/beijing_tianqi_2018.csv"
df = pd.read_csv(fpath)
df.head()
```

```python
df.set_index(df["ymd"])
# 如何删除"ymd"列
df
```

## 复现

```python
# 只选择三月份数据进行分析
condition=df["ymd"].str.startswith("2018-01")
df.loc[condition,"wencha"]=df["bWendu"]-df["yWendu"]
df
```

```python
df[condition]["Wencha_2"]=df["bWendu"]-df["yWendu"]
```

### 警告原因
链式操作其实是两个步骤，先get然后set，get得到的dataframe可能是view也可能是copy。   
解决方法：
+ get+set的两步操作，改成set的一步操作
+ 使用copy复制dataframe df_2=df\[condition\].copy() 先copy得到

### 总结
pandas不允许删选子dataframe，再进行修改写入。
+ 要么使用.loc实现一个步骤的直接修改源
+ 先复制一个子dataframe再进行修改
