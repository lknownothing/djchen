### Pandas实现数据的合并concat
给dataFrame添加行，给dataframe添加列   
一句说明concat语法：
+ 使用某种合并方式（inner/outer)
+ 沿着某个轴向（axis=0/1)
+ 把多个dafaframe合并成一个    

concat语法：pandas.concat(objs,axis=0,join="outer",ingnore_index=False)
+ objs：一个列表，内容可以时DataFrame
+ axis: 默认是按照行合并，如果等于1代表按列合并
+ join 合并的时候，索引的对齐方式，默认是outer join
+ ingore index:是否忽略原来的数据索引

append语法：dataframe.append(other,ignore_index=False)
+ 只有按行合并

```python
import pandas as pd
df1=pd.DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
     'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3'],
     'E': ['E0', 'E1', 'E2', 'E3']
})
df1
```

```python
df2 = pd.DataFrame({ 'A': ['A4', 'A5', 'A6', 'A7'],
                     'B': ['B4', 'B5', 'B6', 'B7'],
                     'C': ['C4', 'C5', 'C6', 'C7'],
                     'D': ['D4', 'D5', 'D6', 'D7'],
                     'F': ['F4', 'F5', 'F6', 'F7']
                   })
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],ignore_index=True)  # 忽略掉原来的索引 新建一个
```

```python
# 以inner的方式concat
pd.concat([df1,df2],ignore_index=True,join="inner")
```

```python
s1 = pd.Series(list(range(4)), name="F")
pd.concat([df1,s1], axis=1)  # axis=1相当于新加一列
```
