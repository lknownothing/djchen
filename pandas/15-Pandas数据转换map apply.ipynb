### 数据转换函数:map apply apply map
+ map 只用于series实现key->value的对应
+ apply:用于series实现每个值得处理，用于dataframe实现某个轴的series处理
+ applymap: 只能用于dataframe，用于处理该dataframe的每个元素

### map用于series值得转换
####  方法一：series.map(dict)
实现：
+ 将股票中的英文名字换成中文名字

```python
import pandas as pd
stocks=pd.read_excel("./datas/stocks/互联网公司股票.xlsx")
stocks.head()
```

```python
# 公司股票代码到中文的映射
dict={
    "bidu":"百度",
    "baba":"阿里巴巴",
    "jd":"京东",
    "iq":"爱奇艺"
}
#　map函数实现变更对应位置的名称
stocks["中文名字"]=stocks["公司"].str.lower().map(dict)  # map实现键值的变换
```

```python
stocks.head()
```

#### 方法二： series.map(function)
function的参数是series的每个元素的值

```python
stocks.loc[:,"中文名字2"]=stocks["公司"].map(lambda x : dict[x.lower()])
# x是传递过来的
```

### 2. apply用于series和dataframe
+ series.apply(function)函数的参数是每个值
+ dataframe.apply(function) 函数的参数是series

#### 2.1 series应用apply(function)

```python
# series应用apply(function)
stocks["中文名字3"]=stocks["公司"].apply(
    lambda x : dict[x.lower()])
stocks.head()
```

```python
stocks.loc[:,["公司","开盘"]]
```

#### 2.2 dataframe.apply(function)
+ function的参数是对应轴的series

apply是在stocks这个dataframe上调用；
lambda x的x是一个series,因为指定了axis=1，所以series的key是列名，可以用x\["公司"\]获取

```python
stocks["公司中文4"]=stocks.apply(
lambda x : dict[x["公司"].lower()],
    axis=1
)
stocks.head()
```
