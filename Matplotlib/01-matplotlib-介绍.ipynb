```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(-10,10,.01)
y=np.sin(x)
plt.plot(x,y)
plt.ylabel("sin(x)")
plt.xlabel("entry x")
```

```python
x=np.linspace(-3,3,1000)
y=np.cos(x)
plt.axis([-5,10,-5,10])
plt.plot(x,y)
plt.show()
```

```python
x=np.linspace(0.00001,1,1000)
y=-x*np.log(x)
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,x*x,x,x**3,x,np.sqrt(x))
plt.show()
```
